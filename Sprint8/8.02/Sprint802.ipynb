#### <span style="color: rgb(230, 0, 126);"> Exercici 5:  Dues variables categ√≤riques. </span>

```python
# El c√≥digo siguiente, que crea un dataframe y quita las filas duplicadas, siempre se ejecuta y act√∫a como un pre√°mbulo del script: 

# dataset = pandas.DataFrame(id, declined, country, year)
# dataset = dataset.drop_duplicates()

# Pegue o escriba aqu√≠ el c√≥digo de script:

import seaborn as sns
import matplotlib.pyplot as plt

dataset = dataset[dataset['year'] == 2021]

# Comptar transaccions per pa√≠s i declined
declined_accepted_by_country = dataset.groupby(['country', 'declined']).size().unstack(fill_value=0) #aix√≤ tamb√© es pot fer amb crosstab

# Reanomenar columnes
column_map = {0: 'Acceptades', 1: 'Rebutjades'}
declined_accepted_by_country = declined_accepted_by_country.rename(columns=column_map)

# Filtrar pa√Øsos amb m√©s de 10 transaccions
declined_accepted_by_country['Total'] = declined_accepted_by_country['Acceptades'] + declined_accepted_by_country['Rebutjades']
declined_accepted_by_country = declined_accepted_by_country[declined_accepted_by_country['Total'] > 10]

# Ordenar per total descendent
declined_accepted_by_country = declined_accepted_by_country.sort_values(by='Total', ascending=False)

# Fem un gr√†fic de columnes apilades 
colors = ['grey', 'steelblue'] 
declined_accepted_by_country_barplot = declined_accepted_by_country[['Rebutjades', 'Acceptades']].plot(kind='bar', stacked=True, figsize=(6,4), color=colors)

# Afegir text a cada part de la barra
for i, (rej, acc) in enumerate(zip(declined_accepted_by_country['Rebutjades'], declined_accepted_by_country['Acceptades'])):
    # Text rebutjades (a la part inferior)
    if rej > 0:
        declined_accepted_by_country_barplot.text(i, rej / 2, str(rej), ha='center', va='center', color='white', fontweight='bold')
    # Text acceptades (a la part superior)
    if acc > 0:
        declined_accepted_by_country_barplot.text(i, rej + acc / 2, str(acc), ha='center', va='center', color='white', fontweight='bold')

plt.title('Transaccions acceptades i rebutjades per pa√≠s (>10 transaccions)')
plt.xlabel('Pa√≠s')
plt.ylabel('Nombre de transaccions')
plt.xticks(rotation=45)
plt.legend(title='Estat')
plt.tight_layout()
plt.show()
```

```python
# El c√≥digo siguiente, que crea un dataframe y quita las filas duplicadas, siempre se ejecuta y act√∫a como un pre√°mbulo del script: 

# dataset = pandas.DataFrame(id, product_id, transaction_id, year, declined, age, user_id, amount)
# dataset = dataset.drop_duplicates()

# Pegue o escriba aqu√≠ el c√≥digo de script:

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dataset = dataset[dataset['year'] == 2021]
dataset = dataset[dataset['declined'] == 0]

# Primera matriu 
products_per_transaction_count = dataset.groupby(['transaction_id'])['product_id'].count().reset_index(name='number_items')
transactions_corr = pd.merge(dataset, products_per_transaction_count, how="left", left_on='id', right_on='transaction_id')
transactions_corr = (transactions_corr.groupby(['user_id','age'])[['amount','number_items']].mean().reset_index())

transactions_corr_matrix = transactions_corr.drop(columns = "user_id").corr(method="spearman")

rename_map = {
    'amount': 'Despesa mitjana ($)',
    'age': 'Edat client',
    'number_items': 'Nombre mitj√† √≠tems'
}

df_renamed = transactions_corr_matrix.rename(columns=rename_map, index=rename_map)

sns.heatmap(df_renamed, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriu de correlaci√≥ entre edat dels clients, \n despesa mitjana ($) i nombre mitj√† items per transacci√≥')
plt.tight_layout()
plt.show()
```

#### <span style="color: rgb(230, 0, 126);"> Exercici 1 (part 2):  </span>

```python
# El c√≥digo siguiente, que crea un dataframe y quita las filas duplicadas, siempre se ejecuta y act√∫a como un pre√°mbulo del script: 

# dataset = pandas.DataFrame(id, product_id, transaction_id, price, declined, year, weight)
# dataset = dataset.drop_duplicates()

# Pegue o escriba aqu√≠ el c√≥digo de script:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# 1. Get product info (one row per product)
product_info = dataset[['product_id', 'weight', 'price']].drop_duplicates(subset='product_id')

# 2. Get total units sold per product (valid transactions only)
df = dataset[(dataset['year'] == 2021) & (dataset['declined'] == 0)]
units = df.groupby('product_id').size().reset_index(name='total_units')

# 3. Merge into a single DataFrame ‚Äî some products may have NaN in total_units
merged = pd.merge(product_info, units, on='product_id', how='left')

# 4. Make sure all are numeric
for col in ['weight', 'price', 'total_units']:
    merged[col] = pd.to_numeric(merged[col], errors='coerce')

# 5. Prepare correlation matrices
cols = ['weight', 'price', 'total_units']
corr_matrix = pd.DataFrame(index=cols, columns=cols, dtype=float)
n_matrix = pd.DataFrame(index=cols, columns=cols, dtype=int)

# 6. Pairwise correlation using max available data per pair
for col1 in cols:
    for col2 in cols:
        pair = merged[[col1, col2]].dropna()
        n = len(pair)
        if n >= 2:
            corr_result = spearmanr(pair[col1], pair[col2])
            corr = corr_result.correlation if not isinstance(corr_result.correlation, np.ndarray) else np.nan
        else:
            corr = np.nan
        corr_matrix.loc[col1, col2] = corr
        n_matrix.loc[col1, col2] = n

# 7. Plot
corr_matrix = corr_matrix.astype('float')
n_matrix = n_matrix.astype('float')

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[0])
axes[0].set_title('Correlation Matrix (Spearman, pairwise)')

sns.heatmap(n_matrix, annot=True, cmap='Greys', fmt=".0f", ax=axes[1])
axes[1].set_title('Number of Observations (N per pair)')

plt.tight_layout()
plt.show()
```

```python
# El c√≥digo siguiente, que crea un dataframe y quita las filas duplicadas, siempre se ejecuta y act√∫a como un pre√°mbulo del script: 

# dataset = pandas.DataFrame(id, amount, country, year, declined)
# dataset = dataset.drop_duplicates()


# Pegue o escriba aqu√≠ el c√≥digo de script:
import seaborn as sns
import matplotlib.pyplot as plt

dataset = dataset[dataset['year'] == 2021]
dataset = dataset[dataset['declined'] == 0]

# Calculem les dades que volem protejar
agg_df = dataset.groupby('country').agg(
    total_transactions=('id', 'count'),
    total_amount=('amount', 'sum'),
    mean_amount=('amount', 'mean'),
    std_amount=('amount', 'std')
).reset_index()

# Per fer el plot necessitem aquestes dades en long format
plot_df = agg_df.melt(
    id_vars='country',
    value_vars=['total_transactions', 'total_amount', 'mean_amount'],
    var_name='metric',
    value_name='value'
)

# Fem el plot
g = sns.FacetGrid(data=plot_df, col='metric', sharey=False, height=4, aspect=1.5)

def barplot_with_error(data, **kwargs): 
    metric = data['metric'].iloc[0]
    if metric == 'mean_amount':
        # Merge std back in
        stds = agg_df[['country', 'std_amount']]
        data = data.merge(stds, on='country')
        sns.barplot(data=data, x='country', y='value', yerr=data['std_amount'], **kwargs)
    else:
        sns.barplot(data=data, x='country', y='value', **kwargs)

g.map_dataframe(barplot_with_error)
g.set_axis_labels("Country", "")
g.set_titles("{col_name}")
g.set_xticklabels(rotation=45)

plt.tight_layout()
plt.show()
```
